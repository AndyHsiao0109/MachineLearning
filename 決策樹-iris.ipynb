```python
from sklearn.datasets import load_iris #從sklearn的數據庫匯入iris的資料
from sklearn.model_selection import train_test_split #從sklearn的model_selection，匯入train_test_split函數(用於分割訓練與測試資料)
from sklearn import tree #從sklearn匯入決策樹分類器

#讀入鳶尾花資料
iris=load_iris()
iris_data=iris.data
iris_label=iris.target

#切分訓練與測試資料
train_data, test_data, train_label, test_label = train_test_split(iris_data, iris_label, test_size = 0.3)

#建立分類器
clf=tree.DecisionTreeClassifier()
clf.fit(train_data, train_label)

print("訓練資料正確率:",clf.score(train_data,train_label)) #訓練資料正確率
print("測試資料正確率:",clf.score(test_data, test_label)) #測試資料正確率

```

```python
import graphviz #匯入graphviz(決策樹可視化)
dot_data=tree.export_graphviz(clf, out_file=None) 
graph=graphviz.Source(dot_data) 
graph.render("iris") #將結果存成pdf檔案
     
```

```python
import pydotplus  #匯入pydotplus模塊
from IPython.display import Image  #從IPython中的display函數，匯入Iamge
dot_data=tree.export_graphviz(clf, out_file=None, 
                feature_names=iris.feature_names,
                class_names=iris.target_names,  
                filled=True, rounded=True,  
                special_characters=True)  
graph=pydotplus.graph_from_dot_data(dot_data)  
Image(graph.create_png())  #將結果用圖形化呈現
```
